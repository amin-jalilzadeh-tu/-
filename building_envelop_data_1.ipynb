```python
import pandas as pd

def dict_to_dataframe(data_dict):
    """
    Flattens the nested building envelope dictionary into a list of records,
    then converts those records to a pandas DataFrame.
    """
    records = []

    # 1) Loop over each building type
    for building_type, year_ranges in data_dict.items():
        # 2) For each year range
        for year_range, scenarios in year_ranges.items():
            # 3) For each scenario
            for scenario_key, elements in scenarios.items():
                # 4) For each building element (e.g., ground_floor, windows, doors, etc.)
                for element_key, element_data in elements.items():
                    # Prepare a record (row) with relevant keys
                    row = {
                        "building_type": building_type,
                        "year_range": year_range,
                        "scenario": scenario_key,
                        "element": element_key,
                        "area_m2": element_data.get("area_m2", None)
                    }

                    # Handle R_value_range if present
                    if "R_value_range" in element_data:
                        r_min, r_max = element_data["R_value_range"]
                        row["R_value_min"] = r_min
                        row["R_value_max"] = r_max
                    else:
                        row["R_value_min"] = None
                        row["R_value_max"] = None

                    # Handle U_value_range if present
                    if "U_value_range" in element_data:
                        u_min, u_max = element_data["U_value_range"]
                        row["U_value_min"] = u_min
                        row["U_value_max"] = u_max
                    else:
                        row["U_value_min"] = None
                        row["U_value_max"] = None

                    # Append the record to the list
                    records.append(row)

    # Convert the list of records to a DataFrame
    df = pd.DataFrame(records)
    return df


df = dict_to_dataframe(building_envelop_data)
print(df)
df.to_csv("building_envelop_data_4.csv", index=False)
```
